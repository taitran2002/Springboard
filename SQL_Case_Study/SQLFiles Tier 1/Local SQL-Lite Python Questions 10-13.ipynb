```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def question10(conn):
    '''Produce a list of facilities with a total revenue less than 1000
    '''
    print("\n Question 10 Answer: ")
    cur = conn.cursor()
#     query = """
#         SELECT name, sub.revenue
#         FROM (
#             SELECT f.name,
#                 SUM(CASE WHEN b.memid = 0 THEN b.slots*f.guestcost
#                     ELSE b.slots*f.membercost END) AS revenue
#                 FROM Facilities AS f
#                 INNER JOIN Bookings AS b
#                 USING (facid)
#                 GROUP BY name) AS sub
#         GROUP BY sub.revenue
#         HAVING sub.revenue < 1000;
#         """
    query = ''' 
    SELECT 
    f.name,
    SUM(CASE WHEN b.memid = 0 THEN b.slots*f.guestcost
        ELSE b.slots*f.membercost END) AS revenue
    FROM Facilities AS f
    INNER JOIN Bookings AS b
    USING (facid)
GROUP BY name
HAVING revenue < 1000;
    '''
    
    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)

def question11(conn):
    '''Produce a report of members and who recommended them in alphabetic surname,firstname order 
    '''
    print("\n Question 11 Answer: ")
    cur = conn.cursor()
    
    query = """
        SELECT sub.surname ||", "|| sub.firstname AS full_name, 
            sub.rec_by
        FROM (SELECT m.firstname,
                m.surname,
                self.surname ||", "|| self.firstname AS rec_by
            FROM Members as m
            LEFT JOIN Members as self
            ON m.recommendedby = self.memid) AS sub
        WHERE sub.rec_by IS NOT NULL
        ORDER BY full_name;
        """
    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)  

def question12(conn):
    '''Find the facilities with their usage by member, but not guests
    '''
    print("\n Question 12 Answer: ")
    cur = conn.cursor()
    
    query = """
        SELECT f.name, 
        COUNT(m.firstname || " " || m.surname)
        FROM Bookings as b
        LEFT JOIN Members as m
        ON m.memid = b.memid
        LEFT JOIN Facilities as f
        ON b.facid = f.facid
        WHERE m.memid <> 0
        GROUP BY f.name
        
        """
    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row) 
        
def question13(conn):
    '''Find the facilities usage by month, but not guests 
    '''
    print("\n Question 13 Answer: ")
    cur = conn.cursor()
    
    query = """
        SELECT f.name, 
        strftime('%m', b.starttime) AS month,
        COUNT(strftime('%m', b.starttime))
        FROM Bookings as b
        LEFT JOIN Members as m
        ON m.memid = b.memid
        LEFT JOIN Facilities as f
        ON b.facid = f.facid
        WHERE m.memid <> 0
        GROUP BY month, f.name
        
        """
    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row) 

        
def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
        question10(conn)
        question11(conn)
        question12(conn)
        question13(conn)
 
 
if __name__ == '__main__':
    main()

```
